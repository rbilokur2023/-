```python
import random

class CardGame:
    CARD_NAMES = ["Dragon", "Knight", "Wizard", "Goblin", "Elf"]

    def __init__(self):
        self.name = random.choice(self.CARD_NAMES)
        self.attack_power = random.randint(5, 15)
        self.health = random.randint(20, 30)

    def attack(self, other):
        other.health -= self.attack_power
        return f"{self.name} –∞—Ç–∞–∫—É—î {other.name}. –ó–¥–æ—Ä–æ–≤'—è {other.name}: {other.health}"

    def defend(self):
        self.health += 2
        return f"{self.name} –∑–∞—Ö–∏—â–∞—î—Ç—å—Å—è. –ó–¥–æ—Ä–æ–≤'—è: {self.health}"

    def make_move(self, other):
        if random.choice([True, False]):
            return self.attack(other)
        else:
            return self.defend()


player_1 = CardGame()
player_2 = CardGame()

print(f"–ì—Ä–∞–≤–µ—Ü—å 1: {player_1.name}, –∞—Ç–∞–∫–∞ {player_1.attack_power}, HP {player_1.health}")
print(f"–ì—Ä–∞–≤–µ—Ü—å 2: {player_2.name}, –∞—Ç–∞–∫–∞ {player_2.attack_power}, HP {player_2.health}")

for i in range(1, 10):
    print(f"\n–†–∞—É–Ω–¥ {i}")
    print(player_1.make_move(player_2))
    print(player_2.make_move(player_1))

    if player_1.health <= 0:
        print(player_1.name, "–ø—Ä–æ–≥—Ä–∞–≤!")
        break
    if player_2.health <= 0:
        print(player_2.name, "–ø—Ä–æ–≥—Ä–∞–≤!")
        break
```

üîπ 6. –ú—ñ–Ω—ñ-–≥—Ä–∞ CardGame

```python
import random

class Item:
    def __init__(self):
        self.number = random.randint(1, 100)
        self.public_attribute = "This is public"
        self._private_attribute = "This is private"
        self.__protected_attribute = "This is protected"

    def __add__(self, other):
        return self.number + other.number

    def __str__(self):
        return f"Item number: {self.number}"

    def __len__(self):
        return self.number % 10

    @property
    def get_protected_attribute(self):
        return self.__protected_attribute


c = Item()
print(c.public_attribute)
c._private_attribute = "Modified private"
print(c._private_attribute)
print(c.get_protected_attribute)
```

üîπ 5. –ü—É–±–ª—ñ—á–Ω—ñ, –∑–∞—Ö–∏—â–µ–Ω—ñ —Ç–∞ –ø—Ä–∏–≤–∞—Ç–Ω—ñ –∞—Ç—Ä–∏–±—É—Ç–∏

```python
import random

class Item:
    def __init__(self):
        self._initial_durability = random.randint(30, 45)
        self.name = random.choice(["trash", "Axe", "Sword", "Spear", "Bow", "Knife"])

    @classmethod
    def modify_class_durability(cls, new_durability):
        obj = cls()
        obj._initial_durability = new_durability
        obj.type = "modified"
        print(f"–°—Ç–≤–æ—Ä–µ–Ω–æ –æ–±'—î–∫—Ç {obj.name} –∑ –º—ñ—Ü–Ω—ñ—Å—Ç—é {obj._initial_durability}")
        return obj


a = Item()
b = Item.modify_class_durability(100)
print(a._initial_durability, b.type)
```

üîπ 4. –ö–ª–∞—Å–æ–≤—ñ –º–µ—Ç–æ–¥–∏ —Ç–∞ –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω—ñ –∫–æ–Ω—Å—Ç—Ä—É–∫—Ç–æ—Ä–∏

```python
import random

class Item:
    global_item_count = 0

    def __init__(self):
        self._initial_durability = random.randint(30, 45)
        self.name = random.choice(["trash", "Axe", "Sword", "Spear", "Bow", "Knife"])

    def test_method(self):
        Item.global_item_count += 1
        return f"{self.name} –º–∞—î –º—ñ—Ü–Ω—ñ—Å—Ç—å {self._initial_durability}"

    @staticmethod
    def always_say_hello():
        return "Hello from Item class!"


i = Item
b = Item()

print(type(i), type(b))
print(b.test_method())
print(i.test_method(b))
print(b.always_say_hello())
print(i.always_say_hello())
```

üîπ 3. –ú–µ—Ç–æ–¥–∏, staticmethod —Ç–∞ self

–ü—Ä–æ–¥–µ–º–æ–Ω—Å—Ç—Ä–æ–≤–∞–Ω–æ —Ä–æ–±–æ—Ç—É –º–µ—Ç–æ–¥—ñ–≤ –∫–ª–∞—Å—É, —Å—Ç–∞—Ç–∏—á–Ω–∏—Ö –º–µ—Ç–æ–¥—ñ–≤ —Ç–∞ –∑–º—ñ–Ω–Ω–æ—ó self.

```python
import random

class Item:
    def __init__(self):
        self._initial_durability = random.randint(30, 45)
        self.name = random.choice(["trash", "Axe", "Sword", "Spear", "Bow", "Knife"])
        self.attack_power = random.randint(4, 6)
        self.durability = self._initial_durability

    @property
    def get_attack_power(self):
        return self.attack_power

    @property
    def get_current_durability(self):
        return f"{self.durability}/{self._initial_durability}"


item = Item()
print(item.attack_power, "=", item.get_attack_power)
item.attack_power += 1
print(item.attack_power, "=", item.get_attack_power)
print("–ú—ñ—Ü–Ω—ñ—Å—Ç—å:", item.get_current_durability)
```

üîπ 2. –í–ª–∞—Å—Ç–∏–≤–æ—Å—Ç—ñ (Property)

–†–µ–∞–ª—ñ–∑–æ–≤–∞–Ω–æ –≤–ª–∞—Å—Ç–∏–≤–æ—Å—Ç—ñ –¥–ª—è –æ—Ç—Ä–∏–º–∞–Ω–Ω—è –ø–æ—Ç–æ—á–Ω–æ—ó –∞—Ç–∞–∫–∏ —Ç–∞ –º—ñ—Ü–Ω–æ—Å—Ç—ñ –ø—Ä–µ–¥–º–µ—Ç–∞.

```python
import random

class Item:
    """–ö–ª–∞—Å –¥–ª—è –æ–ø–∏—Å—É —ñ–≥—Ä–æ–≤–æ–≥–æ –ø—Ä–µ–¥–º–µ—Ç–∞"""

    def __init__(self):
        self.name = random.choice(["trash", "Axe", "Sword", "Spear", "Bow", "Knife"])
        self.rarity = "green"
        self.attack_power = random.randint(4, 6)
        self.durability = random.randint(30, 45)
        self.magic_attribute = "–ú–∞–≥—ñ—è –ü—Ä–∏—Ä–æ–¥–∏"
        self.description = "–†–∞–Ω–¥–æ–º–Ω–∏–π –ø—Ä–µ–¥–º–µ—Ç –∑–±—Ä–æ—ó"

    def attack(self):
        attack = self.attack_power
        if "–ú–∞–≥—ñ—è" in self.magic_attribute:
            attack += 2
        return f"{self.name} –Ω–∞–Ω—ñ—Å {attack} —à–∫–æ–¥–∏ —Ç–∏–ø–æ–º {self.magic_attribute}"

    def defence(self):
        self.durability += 1
        return f"{self.name} –∑–∞—Ö–∏—â–∞—î—Ç—å—Å—è —Ç–∞ –≤—ñ–¥–Ω–æ–≤–ª—é—î –º—ñ—Ü–Ω—ñ—Å—Ç—å –¥–æ {self.durability}"


for i in range(10):
    item = Item()
    print(f"""
–ü–æ–¥—ñ—è {i}:
–Ü—Ç–µ–º: {item.name}
–ê—Ç–∞–∫–∞ / –ú—ñ—Ü–Ω—ñ—Å—Ç—å: {item.attack_power} / {item.durability}
–ú–∞–≥—ñ—è: {item.magic_attribute}
""")
    if random.randint(0, 1):
        print(item.attack())
    else:
        print(item.defence())
```

```python
class Example:

    @classmethod
    def class_info(cls):
        return f"Це клас {cls.__name__}"

    @staticmethod
    def say_hello():
        return "Hello from static method!"


print(Example.class_info())
print(Example.say_hello())
```

## 4. Classmethod та Staticmethod

```python
class User:
    def __init__(self, username):
        self._username = username

    @property
    def username(self):
        return self._username.capitalize()

    @property
    def name_length(self):
        return len(self._username)


u = User("bohdan")
print(u.username)
print("Довжина імені:", u.name_length)
```

### ✅ Завдання 2  
Змініть назву університету для всіх студентів.

```python
class Student:
    university = "IT College Lviv"  # змінна класу

    def __init__(self, name):
        self.name = name  # змінна об'єкта

    def info(self):
        return f"Студент {self.name} навчається в {Student.university}"


s1 = Student("Андрій")
s2 = Student("Олена")

print(s1.info())
print(s2.info())
```

### ✅ Завдання 1  
Створіть ще один об'єкт класу `Person` з іншим ім'ям та віком.

```python
class Person:
    """Клас для опису людини"""

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def say_hello(self):
        return f"Привіт, мене звати {self.name}, мені {self.age} років"


# створення об'єкта
p1 = Person("Богдан", 20)

print(p1.say_hello())
```

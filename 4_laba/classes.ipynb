10. **Магічні методи****

```python
class MyAnimals:
    def __init__(self, animals, name, age):
        self.animals = animals
        self.name = name
        self.age = age

    def __str__(self):
        return f"{self.animals} {self.name} ({self.age})"

    def __repr__(self):
        return f'MyAnimals("{self.animals}", "{self.name}", {self.age})'

    def __len__(self):
        return len(self.name)


a = MyAnimals("Собака", "Рекс", 2)
print(a)
print(repr(a))
print(len(a))
b = MyAnimals("Кошка", "Мурка", 3)
print(b)
```

9. **Приватні атрибути****

```python
class MyAnimals:
    def __init__(self, name):
        self.name = name
        self._secret = "Приватний"
        self.__very_secret = "Дуже приватний"

    @property
    def secret(self):
        return self._secret

    @property
    def very_secret(self):
        return self.__very_secret


a = MyAnimals("Рекс")
print(a.secret)
print(a.very_secret)
```

8. **Методи класу**

```python
class MyAnimals:
    total_objects = 0
    total_csv = 0

    def __init__(self, animals, name, age):
        self.animals = animals
        self.name = name
        self.age = int(age)
        MyAnimals.total_objects += 1

    @classmethod
    def from_csv(cls, data: str):
        cls.total_csv += 1
        animals, name, age = data.split(",")
        return cls(animals, name, age)

    @classmethod
    def check_if_capital(cls, name):
        return name[0].isupper()


a1 = MyAnimals.from_csv("Кіт,Мурка,5")
a2 = MyAnimals("Собака", "Рекс", 2)

print(MyAnimals.check_if_capital(a1.name))
print("Всього:", MyAnimals.total_objects)
print("CSV:", MyAnimals.total_csv)
print("Росомаха.")
```

7. **Properties****

```python
from datetime import datetime

class MyAnimals:
    max_age = 20

    def __init__(self, animals, name, age):
        self.animals = animals
        self.name = name
        self.age = age

    @property
    def live_until(self):
        return datetime.today().year - self.age + self.max_age

    @property
    def see_age(self):
        return f"Моїй тваринці {self.age} років"


a1 = MyAnimals("Кіт", "Мурка", 5)
print(a1.live_until)
print(a1.see_age)
print("Росомаха.")
```

6. **Статичні методи**

```python
class MyAnimals:
    total_animals = 0
    types = set()

    def __init__(self, animals, name, age):
        self.animals = animals
        self.name = name
        self.age = age

        MyAnimals.total_animals += 1
        MyAnimals.types.add(animals)


MyAnimals("Кіт", "Мурка", 5)
MyAnimals("Пес", "Рекс", 3)
MyAnimals("Кролик", "Банні", 2)

print(MyAnimals.total_animals)
print(MyAnimals.types)
print("Росомаха.")
```

5. **Змінні класу****

```python
class MyAnimals:
    total_animals = 0
    total_types = set()

    def __init__(self, animals, name, age):
        self.animals = animals
        self.name = name
        self.age = age

        MyAnimals.total_animals += 1
        MyAnimals.total_types.add(animals)


animals = ["Кіт", "Пес", "Кролик"]
names = ["Bobby", "Perry", "Rex"]

for i in range(3):
    MyAnimals(animals[i], names[i], i)

print("Типи:", MyAnimals.total_types)
print("Кількість:", MyAnimals.total_animals)

print("Росомаха.")

try:
    user_input = int(input("Enter a number: "))
    print(f"You entered: {user_input}")
except ValueError:
    print("Помилка: введено не число")
finally:
    print("Росомаха.")
```

3. **Зміна атрибутів обʼєкта**

```python
class MyAnimals:
    def __init__(self, animals, name, age, mass=None):
        print(f"В мене є тваринка {animals}, яку звати {name}, якій {age} років")
        self.animals = animals
        self.name = name
        self.age = age
        self.mass = 0 if mass is None else mass


a1 = MyAnimals("Рибки", "Неони", 1)
a2 = MyAnimals("Кіт", "Мурка", 5, 10)
a3 = MyAnimals("Собака", "Рекс", 2)

print(a2.name)

a2.name = "Машка"
print(f"Я перейменував тваринку на {a2.name}, вага {a2.mass} кг")

try:
    print("Росомаха...")
except ValueError:
    print("Помилка!")
finally:
    print("Росомаха.")
```

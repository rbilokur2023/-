Отримані знання є базою для подальшого вивчення мови Python.

```python
def привітання():
    return "Привіт, світ!"

def додати(a, b):
    return a + b

def степінь(x, n=2):
    return x ** n

def сума(*числа):
    return sum(числа)

print(привітання())
print(додати(5, 3))
print(степінь(2))
print(степінь(2, 3))
print(сума(1, 2, 3, 4))
```

7. **Функції****



Функції дозволяють перевикористовувати код.

```python
for i in range(5):
    print(i)

фрукти = ["яблуко", "банан", "апельсин"]
for фрукт in фрукти:
    print(фрукт)

лічильник = 0
while лічильник < 3:
    print(лічильник)
    лічильник += 1

квадрати = [x**2 for x in range(5)]
print(квадрати)
```

6. **Цикли**



Цикли дозволяють багаторазово виконувати один і той самий код.

```python
вік = 18
має_паспорт = True

if вік >= 18 and має_паспорт:
    print("Ви можете голосувати")
elif вік >= 18:
    print("Потрібен паспорт")
else:
    print("Ви ще замалі")

статус = "дорослий" if вік >= 18 else "неповнолітній"
print(статус)
```

5. **Умовні конструкції**



Умовні оператори дозволяють виконувати код залежно від умови.

```python
числа = [1, 2, 3, 4, 5]
фрукти = ["яблуко", "банан", "апельсин"]

числа.append(6)
фрукти.extend(["груша", "слива"])

числа.remove(3)
фрукти.pop(1)

числа.sort(reverse=True)
фрукти.sort()

print(числа)
print(фрукти)

print(числа[:2])
print(числа[-2:])
print(числа[::2])
```

4. **Списки та колекції**



Списки дозволяють зберігати набір елементів одного або різних типів.

```python
текст = "Python - чудова мова програмування!"

print(текст)
print(len(текст))
print(текст.upper())
print(текст.lower())

слова = текст.split()
print(слова)
print("-".join(слова))

print(текст.find("мова"))
print(текст.replace("Python", "Python 3"))

print(текст.startswith("Python"))
print(текст.endswith("!"))
print("мова" in текст)
```

3. **Рядки та їх методи****



Рядки — це послідовності символів з багатьма вбудованими методами.

```python
a = 10
b = 3

print(f"Додавання: {a + b}")
print(f"Віднімання: {a - b}")
print(f"Множення: {a * b}")
print(f"Ділення: {a / b}")
print(f"Цілочисельне ділення: {a // b}")
print(f"Залишок від ділення: {a % b}")
print(f"Степінь: {a ** b}")

print("\nПорівняння:")
print(a > b)
print(a < b)
print(a == b)
print(a != b)
```

2. **Основні операції**



Python підтримує всі стандартні математичні та логічні операції

```python
# Числові типи
ціле_число = 42
число_з_комою = 3.14
комплексне_число = 2 + 3j

# Рядковий тип
текст = "Привіт, Python!"

# Логічний тип
істина = True
хибність = False

# Виведемо типи та значення
print(f"Ціле число: {ціле_число}, тип: {type(ціле_число)}")
print(f"Число з комою: {число_з_комою}, тип: {type(число_з_комою)}")
print(f"Комплексне число: {комплексне_число}, тип: {type(комплексне_число)}")
print(f"Текст: {текст}, тип: {type(текст)}")
print(f"Логічне значення: {істина}, тип: {type(істина)}")
```

```python
# =========================
# Основи Python
# =========================

# 1. Змінні та типи даних
ціле_число = 42
число_з_комою = 3.14
комплексне_число = 2 + 3j
текст = "Привіт, Python!"
істина = True
хибність = False

print(f"Ціле число: {ціле_число}, тип: {type(ціле_число)}")
print(f"Число з комою: {число_з_комою}, тип: {type(число_з_комою)}")
print(f"Комплексне число: {комплексне_число}, тип: {type(комплексне_число)}")
print(f"Текст: {текст}, тип: {type(текст)}")
print(f"Логічне значення: {істина}, тип: {type(істина)}")

print("\n" + "="*40)

# 2. Основні операції
a = 10
b = 3

print(f"Додавання: {a + b}")
print(f"Віднімання: {a - b}")
print(f"Множення: {a * b}")
print(f"Ділення: {a / b}")
print(f"Цілочисельне ділення: {a // b}")
print(f"Залишок від ділення: {a % b}")
print(f"Степінь: {a ** b}")

print(f"{a} > {b}: {a > b}")
print(f"{a} == {b}: {a == b}")

print("\n" + "="*40)

# 3. Рядки та їх методи
text = "Python - чудова мова програмування!"

print(text)
print(len(text))
print(text.upper())
print(text.lower())

words = text.split()
print(words)
print("-".join(words))

print(text.find("мова"))
print(text.replace("Python", "Python 3"))

print(text.startswith("Python"))
print(text.endswith("!"))
print("мова" in text)

print("\n" + "="*40)

# 4. Списки та колекції
числа = [1, 2, 3, 4, 5]
фрукти = ["яблуко", "банан", "апельсин"]

print(числа)
print(фрукти)

числа.append(6)
фрукти.extend(["груша", "слива"])

print(числа)
print(фрукти)

числа.remove(3)
фрукти.pop(1)

числа.sort(reverse=True)
фрукти.sort()

print(числа)
print(фрукти)

print(числа[:2])
print(числа[-2:])
print(числа[::2])

print("\n" + "="*40)

# 5. Умовні конструкції
вік = 18
має_паспорт = True

if вік >= 18 and має_паспорт:
    print("Ви можете голосувати")
elif вік >= 18:
    print("Потрібен паспорт")
else:
    print("Замало років")

статус = "дорослий" if вік >= 18 else "неповнолітній"
print(статус)

оцінка = 85
if 90 <= оцінка <= 100:
    рівень = "відмінно"
elif 75 <= оцінка < 90:
    рівень = "добре"
elif 60 <= оцінка < 75:
    рівень = "задовільно"
else:
    рівень = "незадовільно"

print(рівень)

print("\n" + "="*40)

# 6. Цикли
for i in range(5):
    print(f"for: {i}")

лічильник = 0
while лічильник < 3:
    print(f"while: {лічильник}")
    лічильник += 1

for i in range(5):
    if i == 1:
        continue
    if i == 4:
        break
    print(f"i = {i}")

квадрати = [x**2 for x in range(5)]
print(квадрати)

print("\n" + "="*40)

# 7. Функції
def привітання():
    return "Привіт, світ!"

def додати(a, b):
    return a + b

def степінь(x, p=2):
    return x ** p

def сума(*args):
    return sum(args)

print(привітання())
print(додати(5, 3))
print(степінь(2))
print(степінь(2, 3))
print(сума(1, 2, 3, 4))

def факторіал(n):
    if n <= 1:
        return 1
    return n * факторіал(n - 1)

print(факторіал(5))

print("\n" + "="*40)

# 8. Lambda
square = lambda x: x * x
print(square(6))

print("\n" + "="*40)

# 9. try / except / finally
try:
    print(10 / 0)
except Exception as e:
    print("Помилка:", e)
finally:
    print("Кінець блоку try")

print("\n" + "="*40)

# 10. Робота з файлом
with open("test.txt", "w") as f:
    f.write("Python is awesome")

with open("test.txt", "r") as f:
    print(f.read())
```
